```python
ids = []
maxID = -1
for line in data.split("\n"):
    rowLow = 0
    rowHigh = 127
    for c in line[:7]:
        if (c == "F"):
            rowHigh =math.floor((rowHigh + rowLow) / 2)
        else:
            rowLow = math.ceil((rowHigh + rowLow) / 2)
    row = rowLow
    seatLow = 0
    seatHigh = 7
    for c in line[7:10]:
        if (c == "L"):
            seatHigh =math.floor((seatHigh + seatLow) / 2)
        else:
            seatLow = math.ceil((seatHigh + seatLow) / 2)
    seat = seatLow
    seatID = row * 8 + seat
    ids.append(seatID)
    if (seatID > maxID):
        maxID = seatID
print(maxID)
```
